```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

def get_job_listings():
    url = 'https://www.linkedin.com/jobs/search/?currentJobId=3856081734&f_C=79295228&f_CR=103323778&geoId=92000000'
    response = requests.get(url)
    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')
        job_listings = soup.find_all('div', {'class': 'job-search-card'})
        
        job_data = []
        for job in job_listings:
            title = job.find('h3', {'class': 'base-search-card__title'}).text.strip()
            company = job.find('a', {'class': 'hidden-nested-link'}).text.strip()
            location = job.find('span', {'class': 'job-search-card__location'}).text.strip()
            anchor_tag = job.find('a', class_='base-card__full-link')
            href_link = anchor_tag['href']
            job_data.append({'Title': title, 'Company': company, 'Location': location, 'Job Link': href_link})
        num_jobs = len(job_data)
        df = pd.DataFrame(job_data)
        return num_jobs,df
    else:
        print("Failed to fetch job listings.")
        return None, None

# Llama a la función para obtener el DataFrame de ofertas de trabajo
num_jobs,df_job_listings = get_job_listings()
if df_job_listings is not None:
    print("Cantidad de empleos:", num_jobs)
    print(df_job_listings.head())  # Muestra las primeras filas del DataFrame
```
